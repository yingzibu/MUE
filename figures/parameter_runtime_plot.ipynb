```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate pseudodata
tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4', 'Task 5']
# parameters = np.random.randint(100, 1000, size=5)
parameters = np.array([3823507, 3823808, 3824109, 3824410, 3824711])
mean_runtimes = np.array([0.340, 0.347, 0.357, 0.364, 0.364])
# mean_runtimes = np.random.randint(10, 100, size=5)
# std_runtimes = np.random.randint(1, 10, size=5)
std_runtimes = np.array([0.007, 0.009, 0.012, 0.025, 0.011])

# Reshape data for plotting
parameter_data = {
    'Task': tasks,
    'Parameter': parameters
}
runtime_data = {
    'Task': tasks,
    'Mean Runtime (ms)': mean_runtimes,
    'Std Dev (ms)': std_runtimes
}
parameter_df = pd.DataFrame(parameter_data)
runtime_df = pd.DataFrame(runtime_data)

# Plot using Seaborn
fig, axes = plt.subplots(ncols=2, figsize=(12, 6), sharey=True)

# Plot for Parameter
sns.barplot(x='Parameter', y='Task', data=parameter_df, ax=axes[0], color='skyblue')
axes[0].set_title('Parameter vs Task')
axes[0].set_xlabel('Number of Parameters')

for i, p in enumerate(parameters):
    axes[0].text(p, i, str(p), ha='right', va='center', color='black')


# Plot for Runtime
sns.barplot(x='Mean Runtime (ms)', y='Task', data=runtime_df, ax=axes[1], color='salmon', alpha=0.7)
axes[1].errorbar(x=runtime_df['Mean Runtime (ms)'], y=runtime_df['Task'], xerr=runtime_df['Std Dev (ms)'], fmt='o', color='purple')
axes[1].set_title('Runtime vs Task')
axes[1].set_xlabel('Mean Runtime (ms)')


plt.suptitle('Parameter and runtime comparison for single and multi-task, Attentive FP', fontsize=16, y=1)

plt.tight_layout()

plt.show()
```
